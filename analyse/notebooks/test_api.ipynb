```python
import sys
import os
import json
from imilo import MiloClient, extract

try:
    base_url = "https://api-ods.dsiml.org"  
    client = MiloClient(base_url=base_url, token="")
    token = client.get_token()
    
    if not token:
        print("Failed to get token")
        sys.exit(1)
    
    structures_bytes = extract(
        id='structures',
        url=base_url,  
        token=token
    )
    
    structures = json.loads(structures_bytes.decode())
    
    # Save to a file
    with open('structures.json', 'w') as f:
        json.dump(structures, f, indent=2)
        
    print(f"Successfully saved structures to structures.json")
    
except Exception as e:
    print(f"Error: {str(e)}")
```
